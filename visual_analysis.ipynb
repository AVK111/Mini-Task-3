```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"crocodile_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
print("Dataset shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())
print("\nBasic stats:\n", df.describe())
```

```python
# --- Bar Plots (Categorical Columns) ---
for col in ["Common Name", "Scientific Name"]:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.savefig(f"bar_{col}.png")
    plt.show()
```

```python
# --- Box Plots (Numeric Columns) ---
for col in ["Observed Length (m)", "Observed Weight (kg)"]:
    plt.figure(figsize=(6, 5))
    sns.boxplot(data=df, y=col)
    plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.savefig(f"box_{col}.png")
    plt.show()
```

```python
# --- Violin Plot ---
plt.figure(figsize=(6, 5))
sns.violinplot(data=df, y="Observed Length (m)")
plt.title("Violin Plot - Observed Length")
plt.tight_layout()
plt.savefig("violin_length.png")
plt.show()
```

```python
# --- KDE Plot ---
plt.figure(figsize=(6, 5))
sns.kdeplot(df["Observed Weight (kg)"], fill=True)
plt.title("KDE Plot - Observed Weight")
plt.tight_layout()
plt.savefig("kde_weight.png")
plt.show()
```

```python
# --- Swarm Plot (Length vs Weight) ---
plt.figure(figsize=(7, 6))
sns.scatterplot(data=df, x="Observed Length (m)", y="Observed Weight (kg)", alpha=0.7)
plt.title("Length vs Weight (Scatter / Swarm-like)")
plt.tight_layout()
plt.savefig("swarm_length_vs_weight.png")
plt.show()
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("corr_matrix.png")
plt.show()
```
